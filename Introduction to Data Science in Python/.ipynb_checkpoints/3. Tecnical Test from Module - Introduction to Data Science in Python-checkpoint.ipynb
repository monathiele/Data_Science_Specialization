```python
import pandas as pd
import numpy as np
```

```python
#list of indicators of energy supply and renewable electricity production from the United Nations for the year 2013, 
##### data loading preperation
import xlrd
df = pd.read_excel('Energy Indicators.xls', header=(2), skiprows=15, skipfooter= 38)
df.drop(df.columns[[0, 1]], axis = 1, inplace = True)
df.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
# Convert Energy Supply to gigajoules (Note: there are 1,000,000 gigajoules in a petajoule). 
# For all countries which have missing data (e.g. data with "...") make sure this is reflected as np.NaN values.
df['Energy Supply'] = df['Energy Supply']* 1000000

df['Energy Supply'] = pd.to_numeric(df['Energy Supply'], errors='coerce')
df['Energy Supply per Capita'] = pd.to_numeric(df['Energy Supply per Capita'], errors='coerce')
df['% Renewable'] = pd.to_numeric(df['% Renewable'], errors='coerce')
```

```python
df
```

```python
#There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 
#e.g. 'Bolivia (Plurinational State of)' should be 'Bolivia'. 'Switzerland17' should be 'Switzerland'.

import re

df['Country'] = df['Country'].apply(lambda x: re.sub(r'[0-9]', '', x))
df['Country'] = df['Country'].apply(lambda x: re.sub(r'\(.+?\)\s*', '', x))
```

```python
#Rename the following list of countries 

df['Country'] = df['Country'].replace(
    to_replace=["Republic of Korea", "United States of America","United Kingdom of Great Britain and Northern Ireland", "China, Hong Kong Special Administrative Region"], 
    value=["South Korea", "United States", "United Kingdom","Hong Kong"])
Energy = df
Energy['Country'].unique()

```

```python
#csv containing countries' GDP from 1960 to 2015 from World Bank. 
df2 = pd.read_csv('world_bank.csv', header=4)
df2['Country Name'] = df2['Country Name'].replace(
    to_replace=["Korea, Rep.", "Iran, Islamic Rep.", "Hong Kong SAR, China"], 
    value=["South Korea", "Iran","Hong Kong"])
df2.rename(columns = {'Country Name' : 'Country'}, inplace = True)
GDP = df2
GDP
```

```python
#Load the Sciamgo Journal and Country Rank data for Energy Engineering and Power Technology 
import xlrd
df3 = pd.read_excel('scimagojr-3.xlsx')
ScimEn = df3
ScimEn
```

```python
# merge data
# Join the three datasets: GDP, Energy, and ScimEn into a new dataset (using the intersection of country names). 
# Use only the last 10 years (2006-2015) of GDP data 
# and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15).

data = ScimEn[ScimEn['Rank']<16].merge(Energy, how ='left')
data = data.merge(GDP[['Country','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']], how = 'left')
data.set_index('Country', inplace = True)
```
